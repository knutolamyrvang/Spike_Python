### Metoder for å starte og stoppe bevegelse

Metoden start() vil skru på motorene til den farten (og retningen) som blir gitt.
Disse start-metodene har ikke noen varighet eller distanse. Etter at motorene er satt på, går programmet til neste kodelinje. Motorene vil fortsette å gå helt til de blir stoppet eller kontrollert av en annnen metode. Metoden stop() vil stoppe motorene uansett hva de holder på med. Det finnes også metoder hvor du kan kontrollere kraften på motorene i stedet for farten. Under finner du metodene:


### motor_pair.start_tank()
Denne metoden:
- Starter motorene med en gitt kraft (for venstre og høyre motor)
- Kraften kan settes hvor som helst mellom -100% og 100%
- Denne metoden kalles asynkron, fordi den planlegger at motorene kjører i bakgrunnen
- Motorene vil kjøre inntil de stoppes av .stop()-metoden

Eksempel på kode:

```python
motor_pair.start_tank(50,50)
```

### wait/sleep

Siden start() og stop() er asynkrone, så trenger de å bli brukt sammen med annen kode for å være nyttig. En vanlig måte er å bruke vente-funksjoner, for eksempel skal programmet vente 5 sekunder.

Det er to måter å bruke vente-funksjoner for varighet:

1. Bruk standard Python API. Sett import time ved begynnelsen av programmet ditt. Bruk metoden time.sleep(seconds) for å vente en viss varighet.

2. Bruk LEGO API. Sett wait_for_seconds(seconds) hvor som helst i programmet ditt
