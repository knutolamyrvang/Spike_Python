### Konfigurere koden til roboten

Det er viktig å konfigurere koden slik at den passer til roboten. Med konfigurering av koden mener jeg de faste elementene som er med i <em>alle</em> programmene du skal lage.

Før du kan programmere roboten er det viktig å huske på følgende:
    - Hvilke porter er motorene koblet til?
    - Hvilken størrelse på hjulene bruker du?
    - Hvor fort vil du kjøre?
    - Hvordan skal roboten stoppe?
Disse tingene skal være med i alle programmene dine.

## Sette opp en kjøring

Før du kan bruke metodene for kjøring, så må du sette opp hva som er dine motorpar. Dette gjør man med følgende funksjon:

MotorPair(left_motor_port, right_motor_port)

Funksjonen over setter opp hva som er venstre og høyre hjul, og her setter du inn portene som motorene på din robot er koblet til.

Funksjonen MotorPair har tre metoder som vi skal bruke videre: 

1. set_motor_rotation(amount, unit='cm').
Denne setter opp hvor langt roboten kjører på 1 rotasjon.

2. set_default_speed(speed)
Her setter man opp hva som er standard fart for roboten i programmet

3. set_stop_action(action)
Dette bestemmer måten roboten skal stoppe på (bremse, holde posisjonen eller flyte)

### Bruke MotorPair
For å bruke funksjonen MotorPair, så må begge motorene i motorparet bli satt opp.  Hvis du f.eks har venstre motor koblet til port A og høyre til port E, så bruker du følgende kode:

```python
motor_pair = MotorPair('A', 'E')
```

### Hjulstørrelse

Metoden som kalles move beveger roboten i den distansen du vil kjøre:

```python
motor_pair.move(amount, unit='cm', steering= 0, speed= None)
```

Men før du kan bruke denne må du sette inn størrelsen på hjulene inn i amount. Om du bruker de små hjulene, så er denne på 17,6 cm og de store hjulene er 27,6 cm. Kodene under viser hvordan man skal sette opp begge:

```python
motor_pair.set_motor_rotation(17.6, 'cm')
motor_pair.set_motor_rotation(27.6, 'cm')
```

### Måter å stoppe på

'brake' - etter kjøring stopper motorene brått ved låsing av hjul og skrur av motorene. Du kan bevege på hjulene med                 hånden.
'hold' - etter kjøring stopper motorene brått, men bruker kraft til å holde hjulene låst. Du kan ikke bevege på hjulene med          hånden din.
'coast' - etter kjøring, skrur av motorene. Om roboten har stor fart vil den fortsette å bevege seg til den stopper

Slik vil kodene se ut:

```python
motor_pair.set_stop_action('brake')
motor_pair.set_stop_action('hold')
motor_pair.set_stop_action('coast')
```

### Sette standard fart
Om ingen verdi for fart blir angitt til move-metoden, vil den bruke standard fart. For eksempel vil koden under bruke 30 i fart, siden ingen annen fart er angitt og fordi 30 er satt som standard(default)


```python
motor_pair = MotorPair('A', 'E')
motor_pair.set_stop_action('hold')
motor_pair.set_motor_rotation(17.5, 'cm')
motor_pair.set_deafult_speed(30)
motor_pair.move(15, 'cm', steering= 0)
```
